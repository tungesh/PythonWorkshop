# Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as scp

# For presentation purposes.
%matplotlib inline    
```

Let's start by plotting three points: (1, 4), (2, 3), and (3, 6).

```python
x = [1, 2, 3, 5]    # List of x coordinates.
y = [4, 3, 6, 2]    # List of y coordinates.
plt.plot(x, y)
plt.show()       # Displays the plot.
plt.clf()        # Clears the plot.
```

The plot command takes two lists as arguments. The first is a list of all all of the x coordinates of the points, and the second a list of the y coordinates.

However this plot isn't what we want. Matplotlib has gone ahead and interpolated a line between consecutive points. We want just the three points, not the line between them. Let's fix this.

```python
plt.plot(x, y, ls='None')    #ls: Linestyle
plt.show()
plt.clf()
```

Looks like we've gone to far. We got rid of the line but now have nothing at all.

```python
plt.plot(x, y, ls='None', marker="D")    # D is for diamond
plt.show()
plt.clf
```

Now we've got one point. The other two are hidden away at the edge. Let's move the boundaries of the plot.

```python
plt.plot(x, y, ls='None', marker="D")    # D is for diamond
plt.xlim([0,4])    # Displays the x-axis from 0 to 4
plt.ylim([0,7])    # Displays the y-axis from 0 to 7
plt.show()
plt.clf
```

Finally we've got the three points drawn. This is a pretty boring plot. Let's try plotting a sin curve.

```python
x = np.linspace(-np.pi, np.pi, 10, True)
y = np.sin(x)
plt.plot(x, y, ls='None', marker=".")
plt.ylim([-1.1, 1.1])
plt.show()
plt.clf()
```

A whole bunch of stuff just happened. Let's take a minute to go through it:
* **np.pi**: The numpy library has a value called pi, which conveniently stores the value of pi.
* **np.linspace**: This helpful function is used to generate a list of 100 points between -pi and pi. The true is used to include the endpoint. Read all about it [here](http://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html).
* **np.sin**: This is a sin function provided by the numpy library.

An important thing to wrap your head around is that plotting in Python is all about points. To plot a function, you evaluate the function at many points and then plot those points. Let's give it a shot.

```python
x = np.linspace(-np.pi, np.pi, 10, True)
y = np.sin(x)
plt.plot(x, y)
plt.ylim([-1.1, 1.1])
plt.show()
plt.clf()
```

This doesn't look particularly smooth. We can fix that be increasing the number of points the function is evaluated at.

```python
x = np.linspace(-np.pi, np.pi, 100, True)
y = np.sin(x)
plt.plot(x, y)
plt.ylim([-1.1, 1.1])
plt.show()
plt.clf()
```

That looks much smoother. Now let's plot multiple figures in a single plot.

```python
x = np.linspace(-np.pi, np.pi, 100, True)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.ylim([-1.1, 1.1])
plt.show()
plt.clf()
```

As simple as making two calls two plot. However with two different things on the plot it's a good idea to label them. Let's add a plot title, axis labels and grid whil were at it.

```python
x = np.linspace(-np.pi, np.pi, 100, True)
plt.plot(x, np.sin(x), label="Sin(x)")
plt.plot(x, np.cos(x), label="Cos(x)")
plt.ylim([-1.1, 1.1])
plt.legend(loc='upper left')    # Display the legend in the upper right.
plt.xlabel("x") # Text for the x-axis.
plt.ylabel("y") # Text for the y-axis.
plt.grid() # Add a grid to the plot.
plt.show()
plt.clf()
```

And that's the general gist of generating plots.
